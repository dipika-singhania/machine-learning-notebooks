8. Create a list with the numbers 1 to 10.
 Write a program to loop through the numbers in the list and print “odd” if the number is odd and “even” if the number is even.

```python
#@title Answer8

list_nums = list(range(1, 10))

for num in list_nums:
  if num % 2 == 0:
    print("Num ", num, " is even")
  else:
    print("Num ", num, " is odd")
```
